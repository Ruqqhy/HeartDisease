Splitting and Scaling the data

```python
#Let us define our x and y
x = df.drop('target', axis=1)
y = df['target']
```

```python
#Let us feature scale the columns to be used
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x[['age', 'trestbps', 'chol', 'thalach', 'oldpeak','sex']] = scaler.fit_transform(x[['age', 'trestbps', 'chol', 'thalach', 'oldpeak','sex']])
```

```python
# Let us do our splitting now
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

Modeling and Predictions

```python
# Let us fit our logistics regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
#Let us check our accuracy
from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#Let us use a confusion matrix to check our prediction
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

The confusion_matrix function compares the actual values (y_test) with the predicted values (y_pred) and creates a matrix that summarizes the model's performance. This matrix shows the number of:

True Positives (TP): Correctly predicted 73 positive cases.
True Negatives (TN): Correctly predicted 29 negative cases.
False Positives (FP): Incorrectly predicted 13 positive cases.
False Negatives (FN): Incorrectly predicted 90  negative cases.
